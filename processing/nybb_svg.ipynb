```python
# plot and save as vector (svg)
import geopandas as gpd
import matplotlib.pyplot as plt

nybb = gpd.read_file(gpd.datasets.get_path('nybb')).to_crs(epsg=2263).simplify(0)

fig, ax = plt.subplots(figsize=(10, 10))
nybb.plot(ax=ax, color='#B31B1B', edgecolor='black', linewidth=5)

ax.set_axis_off()

# disable plot background so it saves transparent
ax.set_facecolor('none')
fig.patch.set_facecolor('none')

# save as svg
plt.savefig('nybb.svg', format='svg')
```
